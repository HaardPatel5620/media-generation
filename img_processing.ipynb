```python
### crop
from PIL import Image

def crop_image(input_path, output_path, crop_mode):
    # Open the input image
    im = Image.open(input_path)

    # Determine the dimensions of the input image
    width, height = im.size

    if crop_mode == "portrait":
        # Crop to a portrait aspect ratio (e.g., 4:5)
        new_width = min(width, height * 4 // 5)
        left = (width - new_width) // 2
        top = 0
        right = left + new_width
        bottom = height
    elif crop_mode == "square":
        # Crop to a square aspect ratio
        new_size = min(width, height)
        left = (width - new_size) // 2
        top = (height - new_size) // 2
        right = left + new_size
        bottom = top + new_size
    elif crop_mode == "story":
        # Crop to a story aspect ratio (9:16) for a typical mobile screen
        width, height = im.size
        target_height = height
        target_width = int(target_height * 9 / 16)
        left = (width - target_width) // 2
        top = (height - target_height) // 2
        right = left + target_width
        bottom = top + target_height
    else:
        raise ValueError("Invalid crop mode. Supported modes: portrait, square, story")

    # Perform the crop
    cropped_im = im.crop((left, top, right, bottom))

    # Save the cropped image
    cropped_im.save(output_path, 'JPEG')


# Example usage:
# Crop the input image in portrait mode
path = r'images\flowera.jpg'
crop_image(path, 'output_portrait1.jpg', 'square')

```

```python
### multilinetext using draw.text

from PIL import Image, ImageDraw, ImageFont, ImageOps

def add_multiline_text_to_image(input_image_path, output_image_path, text, position=(10, 10), font_path=None, font_size=20, text_color=(255, 255, 255), max_width=None):
    # Open the image
    image = Image.open(input_image_path)
    draw = ImageDraw.Draw(image)

    # Load the font
    if font_path is None:
        font = ImageFont.load_default()
    else:
        font = ImageFont.truetype(font_path, font_size)

    # Calculate the maximum width for text wrapping
    if max_width is None:
        max_width = image.width - position[0]  # Use the available width

    # Split the text into lines based on word wrapping
    lines = []
    words = text.split()
    current_line = []

    for word in words:
        test_line = current_line + [word] if current_line else [word]
        left, top, right, bottom = font.getbbox(' '.join(test_line))  # Use font.getsize()
        width, height = right - left, bottom - top
        if width <= max_width:
            current_line = test_line
        else:
            lines.append(' '.join(current_line))
            current_line = [word]
    
    if current_line:
        lines.append(' '.join(current_line))

    # Draw the text on the image
    y_position = position[1]
    for line in lines:
        draw.text((position[0], y_position), line, fill=text_color, font=font)
        left, top, right, bottom = font.getbbox(line)  # Use font.getsize()
        width, height = right - left, bottom - top
        y_position += height

    # Save the resulting image
    image.save(output_image_path)

# Example usage:
path = r'images\flowera.jpg'
font_path = r'C:\Windows\Fonts\MAGNETOB'
multiline_text = "This is a a a long multiline text that will be added to the image.\nIt can have multiple lines."
add_multiline_text_to_image(path, 'output_image_with_text1.jpg', multiline_text, position=(10, 10),font_path=font_path, font_size=100, text_color=(255, 255, 255), max_width=300)
```

```python
###text wrap \n 
def text_wrap(text, font, max_width):
    """Wrap text based on the specified width.
    This function adds newline characters to split lines.
    
    @params:
        text: str
            text to wrap
        font: obj
            font of the text
        max_width: int
            width to split the text with
    @return
        wrapped_text: str
            text with added newline characters
    """
    lines = []
    
    # If the text width is smaller than the max_width, then no need to split, just return it
    if font.getlength(text) <= max_width:
        return text
    
    # Split the line by spaces to get words
    words = text.split(' ')
    i = 0
    wrapped_text = ''
    
    while i < len(words):
        line = ''
        while i < len(words) and font.getlength(line + words[i])  <= max_width:
            line = line + words[i] + " "
            i += 1
        if not line:
            line = words[i]
            i += 1
        lines.append(line)
    
    # Join the lines with newline characters
    wrapped_text = '\n'.join(lines)
    return wrapped_text
```

```python
from PIL import Image

def gradient_from_bottom_to_top(path, gradient_magnitude=1.):
    im = Image.open(path)
    if im.mode != 'RGBA':
        im = im.convert('RGBA')
    
    width, height = im.size
    
    gradient = Image.new('L', (1, height), color=0xFF)
    
    for y in range(height):
        # Reverse the gradient by changing the calculation here
        gradient.putpixel((0, y), int(255 * (1 - gradient_magnitude * float(height - y) / height)))
    
    alpha = gradient.resize(im.size)
    
    black_im = Image.new('RGBA', (width, height), color=0) # i.e. black
    black_im.putalpha(alpha)
    
    gradient_im = Image.alpha_composite(im, black_im)
    gradient_im.save('out.png', 'PNG')

# Example usage:
# gradient_from_bottom_to_top('input_image.png', gradient_magnitude=0.5)
```

```python
import cv2
from PIL import ImageFont
import numpy as np

# Load the image
image = cv2.imread(r'images\wall-e.jpg')

# Define the text and font properties
text = "Curved Text"
# font = cv2.FONT_HERSHEY_SIMPLEX

font_path = r'C:\Windows\Fonts\MAGNETOB'
font = ImageFont.truetype(font_path, font_size)

font_scale = 1
font_thickness = 2
font_color = (255, 255, 255)  # White color in BGR

# Create a blank canvas (black) with the same size as the image
height, width, _ = image.shape
canvas = np.zeros((height, width, 3), dtype=np.uint8)

# Create a path for the curved text (e.g., a semi-circle)
center = (200, 200)
radius = 100
angle = -180  # Adjust the angle to control the curvature

start_angle = 180
end_angle = 180 + angle

# start_angle = 180 - angle // 2
# end_angle = 180 + angle // 2
num_points = 11  # Adjust the number of points for smoother curvature

for i in range(num_points):
    theta = np.deg2rad(start_angle + (start_angle -end_angle ) * i / (num_points - 1))
    x = int(center[0] + radius * np.cos(theta))
    y = int(center[1] - radius * np.sin(theta))  # Negative sin to flip the text

    # Render the text at each (x, y) point
    cv2.putText(canvas, text[i % len(text)], (x, y), font, font_scale, font_color, font_thickness)

# Overlay the canvas with curved text on the original image
result = cv2.addWeighted(image, 1, canvas, 1, 0)

# Save the edited image
cv2.imwrite('curved_text_image.jpg', result)
```

```python
import cv2

# Open the input video
input_video = r'videos\testvideo.mp4'
cap = cv2.VideoCapture(input_video)

# Get the video's frame width, height, and frames per second
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = cap.get(5)

# Define the text content and styling
text = "Text Animation"
text_start_time = 2  # Start showing animated text 2 seconds after the video starts
text_font = cv2.FONT_HERSHEY_SIMPLEX
text_font_scale = 1
text_font_color = (0, 0, 0)
text_thickness = 2

# Define the point where the text stops sliding (adjust as needed)
stop_x = frame_width // 2  # Text stops at the center of the frame

# Create a VideoWriter object to save the output video
output_video = 'output_video3.mp4'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video, fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Get the current frame number
    frame_number = int(cap.get(1))

    # Create a blank frame for text animation
    text_frame = frame.copy()

    # Calculate the position for the text animation (right to left)
    if frame_number >= int(fps * text_start_time):
        text_x = int((frame_number - fps * text_start_time) * frame_width / (fps * 3))  # Adjust animation speed here
        if text_x < stop_x:
            cv2.putText(text_frame, text, (text_x, 50), text_font, text_font_scale, text_font_color, text_thickness)
        else:
            cv2.putText(text_frame, text, (stop_x, 50), text_font, text_font_scale, text_font_color, text_thickness)

    # Combine the text frame with the original frame
    combined_frame = cv2.addWeighted(frame, 1, text_frame, 0.5, 0)

    out.write(combined_frame)

# Release video objects
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from moviepy.editor import VideoFileClip, TextClip

# Load the input video
input_video = r'videos\testvideo.mp4'
clip = VideoFileClip(input_video)

# Define the text content, duration, and styling
text = "Text Animation"
text_duration = 5  # Duration of text animation in seconds
text_fontsize = 48
text_color = 'white'

# Create a TextClip for the text animation
text_clip = TextClip(text, fontsize=text_fontsize, color=text_color, size=clip.size)

# Calculate the duration for which the text should be visible
text_start_time = 2  # Start showing text 2 seconds after the video starts
text_end_time = clip.duration  # Show text until the end of the video

# Calculate the keyframes for text visibility
text_appear_frame = clip.duration - (clip.duration - text_duration) / 2

# Create a function to display text at the appropriate keyframes
def add_text_to_frame(t):
    if t < text_start_time:
        return clip.get_frame(t)
    elif t < text_appear_frame:
        return clip.get_frame(text_start_time)
    elif t < text_end_time:
        return clip.get_frame(t).overlay(text_clip.get_frame(t - text_start_time))
    else:
        return clip.get_frame(t)

# Apply the function to the video
video_with_text = clip.fl(add_text_to_frame)

# Write the output video
output_video = 'output_video.mp4'
video_with_text.write_videofile(output_video, codec='libx264', fps=clip.fps)

# Close the video clips
clip.close()
text_clip.close()
```
